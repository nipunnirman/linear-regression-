```python
import numpy as np

def compute_cost(X, Y, beta):
    """
    X : (m , n)
    Y : (m, )
    beta : coefficient vector (including bias term)

    MSE = (1 / 2m) * sum((y_hat - y)^2)
    y_hat = X.dot(beta) => beta0 + beta1*x1 + ... + betaj*xj
    """
    m = len(Y)
    y_hat = X.dot(beta)
    cost = (1 / (2 * m)) * np.sum((y_hat - Y) ** 2)
    return cost


def gradient_descent(X, Y, beta, lr, n_iter):
    """
    dj/dbeta_j = (1/m) * X.T.dot(y_hat - Y)
    """
    m = len(Y)
    cost_history = []

    for i in range(n_iter):
        y_hat = X.dot(beta)
        gradients = (1 / m) * X.T.dot(y_hat - Y)   # gradient formula
        beta = beta - lr * gradients               # update step

        cost = compute_cost(X, Y, beta)
        cost_history.append(cost)

        if i % 100 == 0:
            print(f"Iteration: {i}, Cost: {cost:.4f}")

    return cost_history, beta

 
```
